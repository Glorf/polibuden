```python
import numpy as np
from bayes_opt import BayesianOptimization
from subprocess import check_output, STDOUT

 
pbounds = {
    'd': (-1.0, 1.0),
    'r': (-1.0, 1.0),
    'p': (-1.0, 1.0),
    'd2': (-1.0, 1.0),
    'r2': (-1.0, 1.0),
    'p2': (-1.0, 1.0),
}
def listowy_hyper_param(d, r, p, d2, r2, p2):
    command = ["./optimizer.sh", str(d), str(r), str(p), str(d2), str(r2), str(p2)]
    return -int(check_output(command, stderr=STDOUT).decode().strip())
 
optimizer = BayesianOptimization(
    f=listowy_hyper_param,
    pbounds=pbounds
)
```

```python
optimizer.maximize(
    init_points=10,
    n_iter=200,
    kappa=0.1
)
```
